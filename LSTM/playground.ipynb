```python
from train_seq2seq import generate
from tokenizer import load_tokenizer
import torch
from LSTM_Seq2Seq import *
import json
```

```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
```

```python
with open('configs/config.json', "r") as config_file:
    config = json.load(config_file)
```

```python
tokenizer = load_tokenizer('tokenizer/vocab.txt')
vocab_size = tokenizer.vocab_size

# Initialize encoder-decoder model
encoder = Encoder(vocab_size, config)
decoder = Decoder(vocab_size, config)
model = Seq2Seq(encoder, decoder)
```

```python
model_fp = "checkpoint/lr0.001_ep100_seqsize50_hid256_drop0.1_layers2_best_epoch=4_min_train_loss=5.6721_min_val_loss=5.8345"
model.load_state_dict(torch.load(model_fp)['model_state_dict'])
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu' 
```

```python
test_seq = "which branch should i pull?"
```

```python
model.to(device)
generate(model, tokenizer, test_seq, device, 100)
```
